```python
import pandas as pd
```

```python
df = pd.read_csv('task_dict.csv')
df.head()
```

```python
df.dropna(inplace=True)
df.drop(13,inplace=True)
```

```python
def preprocess(text):
    text = str(text)
    text = text.replace('at any time.,','')
    text = text.replace('\n','')
    return text
```

```python
df['Profession'] = df['Profession'].apply(lambda x: preprocess(x))
df['Name'] = df['Name'].apply(lambda x: preprocess(x))
df['Biography'] = df['Biography'].apply(lambda x: preprocess(x))
```
