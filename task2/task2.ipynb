```python
import pandas as pd
from os import listdir
```

```python
#We can read the scraped files
read_all_1 = [pd.read_excel(f"shohrux/{file}") for file in listdir('shohrux') if file.endswith('.xlsx')]
read_all_2 = [pd.read_excel(f"sirojiddin/{file}") for file in listdir('sirojiddin') if file.endswith('.xlsx')]

df1 = pd.concat(read_all_1, axis=0,ignore_index=True)
df2 = pd.concat(read_all_2, axis=0,ignore_index=True)
```

```python
#Checed df shape
print(f"df1 {df1.shape}")
print(f"df2 {df2.shape}")
```

```python
df1.head()
```

```python
#The necessary columns for us
df1=df1[['LABEL-3','LABEL-9','LABEL-10','LABEL-1']]
df2=df2[['LABEL-3','LABEL-9','LABEL-10','LABEL-1']]
```

```python
#Rename columns
df1.rename(columns={'LABEL-3':'Name','LABEL-9':'Profession','LABEL-10':'Location','LABEL-1':'Image'},inplace=True)
df2.rename(columns={'LABEL-3':'Name','LABEL-9':'Profession','LABEL-10':'Location','LABEL-1':'Image'},inplace=True)
```

```python
#Drop duplicate values
df1.drop_duplicates(inplace=True)
df2.drop_duplicates(inplace=True)
```

```python
# we combine data frames
df_new=pd.concat([df1,df2],axis=0,ignore_index=True)
```
